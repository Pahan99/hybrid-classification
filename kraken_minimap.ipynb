```python
import pandas as pd
import numpy as np
```

```python
filepath = '/content/drive/MyDrive/fyp code/'
```

1. Kraken2 output

```python
kraken_df = pd.read_csv(f'{filepath}/kraken/kraken2.csv')
kraken_df.head()
```

```python
# rename 'name' column
kraken_df.rename(columns={'name':'name_kraken'},inplace=True)
```

2. Minimap2 output

```python
minimap_df = pd.read_csv(f'{filepath}/minimap/combined_minimap.csv')
minimap_df.head()
```

```python
minimap_df.shape
```

```python
# rename 'name' column
minimap_df.rename(columns={'name':'name_minimap'},inplace=True)
```

### Merge outputs

```python
combined_df = pd.merge(minimap_df, kraken_df, on='seq_id', how='left')
```

```python
combined_df.columns
```

```python
combined_df.drop(['status', 'taxonomy_id', 'taxon'], axis=1,inplace=True)
```

```python
combined_df
```

```python
combined_df.to_csv(f'{filepath}/kraken/kraken_minimap.csv',index=None)
```
